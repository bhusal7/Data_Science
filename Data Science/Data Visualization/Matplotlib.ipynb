```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# matplotlib

x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x-Axis")
plt.ylabel('y-Axis')
```

```python
# subplot in matplotlib
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,y**3)
plt.subplot(2,2,4)
plt.plot(x,y)
# lets create some thi g in 4th
```

```python
# object oriented way
fig = plt.figure()
axis1 = fig.add_axes([0.2,0.5,0.9,0.5])
      # [left to right  , buttom to up   ,width , height ] 
axis2 = fig.add_axes([1,1,0.2,0.2])

axis1.plot(x,y)
axis2.plot(x,y)
```

```python
fg = plt.figure(figsize=(20,12), dpi=100)
a = fg.add_axes([1,3,1,1])
b = fg.add_axes([2,0.5,1,1])
a.plot(x,y)
b.plot(x**2,y**3)

# saving plot 
plt.savefig("objectOrientedway.png")
```
