```python
dx = sns.load_dataset('flights')
dx
```

```python
df = sns.load_dataset('tips')
df
```

```python
dfcorr = df[['total_bill','tip','size']]
dfcorr
```

```python
dfcorr.corr()
```

```python
sns.heatmap(dfcorr.corr(),annot=True)
```

```python
sns.clustermap(dfcorr.corr(),annot=True)
```

```python
pv = dx.pivot_table(values='passengers', index='month', columns='year')
pv
```
