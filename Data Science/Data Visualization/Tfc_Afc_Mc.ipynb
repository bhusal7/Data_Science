TFC, TVC, AFC , AVC, MC Curves

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
l = np.arange(0,6)
l
```

```python
k = np.full(6,10)
k
```

```python
q = np.array([1,2,3,4,5,6])
q
```

```python
tvc = np.array([0,20,30,45,80,145])
tvc
```

```python
tfc = np.full(6,100)
tfc
```

```python
df = pd.DataFrame({
    'L': l,
    'K': k,
    'Q': q,
    'Tfc':tfc,
    'Tvc' : tvc
})
df
```

```python
# q=0m tc = tfc
# tvc+tfc=tc
# afc = tfc/q
# avc = tvc/q
# ac = sum of 2 or tc/q
# mc = dy/dx = d(tc)/d(q)
# 
```

```python
df['Tc'] = df['Tfc']+df['Tvc']
df
```

```python
df.columns
```

```python
df['Afc'] = (df['Tfc']/df['Q']).round(2)
df['Avc'] = (df['Tvc']/df['Q']).round(2)
df['Ac'] = df['Afc']+(df['Avc'])
df['Mc'] = df['Tc'].diff()/(df['Q'].diff())
df
```

```python
fig = plt.figure(figsize=(12,10),dpi=100)
plt.title("Curves")
plt.xlabel('Quantity')
plt.ylabel("TFC,TVC,TC,AFC,AVC,AC & MC")
sns.lineplot(x='Q',y='Tfc',data=df,marker="o", label='Tfc')
sns.lineplot(x='Q',y='Tvc',data=df,marker="o", label='Tvc')
sns.lineplot(x='Q',y='Tc',data=df,marker="o", label='Tc')
sns.lineplot(x='Q',y='Afc',data=df,marker="o", label='Afc')
sns.lineplot(x='Q',y='Avc',data=df,marker="o", label='Avc')
sns.lineplot(x='Q',y='Ac',data=df,marker="o", label='Ac')
sns.lineplot(x='Q',y='Mc',data=df,marker="o", label='Mc')
```
