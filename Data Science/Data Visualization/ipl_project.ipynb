```python
# Split bowling figure
df[['wickets', 'runs_given']] = df['best_bowling_figure'].str.split('--', expand=True)

# Convert to int
df['wickets'] = df['wickets'].astype(int)
df['runs_given'] = df['runs_given'].astype(int)

# Sort
best_b = df.sort_values(
    by=['wickets', 'runs_given'],
    ascending=[False, True]
).head(10)

sns.barplot(
    y=best_b['best_bowling'],
    x=best_b['wickets'],
    palette='rocket'
)
```
