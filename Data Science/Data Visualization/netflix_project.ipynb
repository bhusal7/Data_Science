```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df
```

```python

df.head()
```

```python
df= df.dropna(subset=['type','country', 'release_year',	'rating', 'duration'])
df.head()
```

```python
movie_tv = df['type'].value_counts()
sns.barplot(x=movie_tv.index ,y=movie_tv.values,palette='viridis')
plt.title("Movie vs Tv Show")
plt.xlabel('Type')
plt.ylabel("Count")
```

Pie Chart

```python
df.head()
```

```python
rating = df['rating'].value_counts()
fig = plt.figure(figsize=(8,6))
plt.pie(rating,labels=rating.index,autopct='%1.1f%%', startangle=90)
plt.title('Percentage of current ratings')
plt.xlabel('Type')
plt.ylabel("Count")
```

#using histogram, movie duration how much distributed

```python
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'],bins=30,color='red',edgecolor='lime')
plt.title('Movie Duration Histogram')
plt.xlabel('Duration of Minute')
plt.ylabel("No. of Movies")
```

```python
release = df['release_year'].value_counts()
release
sns.scatterplot(x=release.index,y=release.values,color='lime',edgecolor='red')
plt.title('Release year vs No. of shows')
plt.xlabel('No. of shows')
plt.ylabel("Release Year")
```
