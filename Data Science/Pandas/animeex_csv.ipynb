```python
#         Mske a new column for count Episodes

def extract_episode(txt):
    check = False
    data = ''
    for i in txt:
        if i == ')':
            check = False
            
        if i == '(':
            check = True
            continue
        if check == True:
            data += i
            
    return data
   
```
