```python
import numpy as np
import pandas as pd
```

```python
#merge 2 data fdataframe_columns
```

```python
stu = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4],
    'Name': ['Ram', 'Sita', 'Hari', 'Gita'],
    'Class': [12, 12, 11, 10]
}) 
stu
```

```python
mks = pd.DataFrame({
    'Student_ID': [1, 2, 5, 6],
    'Subject': ['Math', 'Science', 'Computer', 'English'],
    'Marks': [85, 90, 78, 88]
})
mks
```

```python
pd.merge(stu,mks)
```

```python
pd.merge(stu,mks,how="outer")
```

```python
pd.merge(stu,mks,on="Student_ID",how='inner')
```

```python
pd.merge(stu,mks,on="Student_ID",how='outer')
```

```python
pd.merge(stu,mks,on="Student_ID",how='left')
```

```python
pd.merge(stu,mks,on="Student_ID",how='right')
```

```python
#Concating
df1 = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': ['Ram', 'Sita', 'Hari', 'Gita'],
    'C': [85, 90, 78, 88]
}) 
df1
```

```python
df2 = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': ['Math', 'Science', 'Computer', 'English'],
    'C': ['Pass', 'Pass', 'Pass', 'Pass']
})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2],axis=1)
```
