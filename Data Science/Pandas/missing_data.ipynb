```python
import numpy as np
import pandas as pd
```

```python
#Missing Data
```

```python
data = {
    'A': [10, 20, 1, 40, 50],
    'B': [np.nan, 15, 25, 35, 45],
    'C': [5, np.nan, 15, 20, np.nan],
    'D': [100, 200, 300, np.nan, np.nan],
    'E': [1, 2, 3, 4, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
#missing value in each column 
df.isna().any()
```

```python
df.isna().any().sum()
```

```python
#removing null
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

```python
#filling the missing ddata
df
```

```python
df.fillna(0)
```

```python
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
values = {'A':10,'B':100,'C':200,'D':300,'E':400}
df.fillna(value=values,inplace=True)
df
```

```python
#fill the mean in placr of nan
df.fillna(df.mean())
df.mean()
```
