```python
import numpy as np 
import pandas as pd 
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("clean_student_dataset.csv")
df.head()

x = df[['daily_study_hours']]
y= df[['exam_score']]

model=LinearRegression()
model.fit(x,y)

predict_score = model.predict(x)

# valid regresssion matrix 
mae = mean_absolute_error(y,predict_score)
mse = mean_squared_error(y,predict_score)
rmse = np.sqrt(mse)
r2 = r2_score(y,predict_score)

print('MAE : On average off by: ',round(mae,2))
print('MSE : Squared Mistake Value: ',round(mse,2))
print('RMSE :Final Realistic Errir: ',round(rmse,2))
print('R^2 score (Model Accuracy): ',round(r2,2))

fig = plt.figure(figsize=(10,6),dpi=100)
fig
sns.histplot(df['exam_score'],color='purple',edgecolor="black",bins=30)
plt.title("Distribution of final exam score")
plt.xlabel("Final Exam Score")
plt.ylabel("No. of Students")
plt.grid(True)
plt.show()

# scatter
fig = plt.figure(figsize=(10,6),dpi=100)
fig
# sns.scatterplot(x,y,color='purple',edgecolor="black",label="Actual Scores")
# sns.regplot(x,predict_score,color='purple',edgecolor="black",label='predict score')

sns.scatterplot(
    x=df['daily_study_hours'],
    y=df['exam_score'],
    color='purple',
    edgecolor='black',
    label='Actual Scores'
)

sns.regplot(
    x=df['daily_study_hours'],
    y=predict_score.flatten(),
    scatter=False,
    color='red',
    label='Prediction Line'
)




plt.title("model prediction vs actual score")
plt.xlabel("Study hrs")
plt.ylabel("final output")
plt.grid(True)
plt.show()
```
