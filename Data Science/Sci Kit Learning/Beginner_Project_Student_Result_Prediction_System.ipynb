```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('students.csv')
df_label = df.copy()
df_label

scaler = StandardScaler()
df_label['Atten_Encode'] = scaler.fit_transform(df_label[['attendance']])
df_label['PrevS_Encode'] = scaler.fit_transform(df_label[['prev_score']])
df_label

x=df_label[['Atten_Encode','PrevS_Encode']]
y=df_label[['passed']]
x_Train,x_Test,y_Train,y_Test = train_test_split(x,y,test_size=0.2,random_state=42)
print(f'Xtrain: {x_Train} \t xtest: {x_Test} \n ytrain: {y_Train} \t ytest: {y_Test}  ')


model = LogisticRegression(x,y)
model.fit()


model.predict([[]])

```
