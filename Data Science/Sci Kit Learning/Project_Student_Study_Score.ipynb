```python
import numpy as np 
import pandas as pd 
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression

df = pd.read_csv('student_study_score.csv')

x = df[['study_hours']]
y=df[['score']]

model = LinearRegression()
model.fit(x,y)

predict_score = model.predict(x)
mae = mean_absolute_error(y,predict_score)
mse = mean_squared_error(y,predict_score)
rmse = np.sqrt(mse)

print('MAE : On average off by: ',mae)
print('MSE : Squared Mistake Value: ',mse)
print('RMSE :Final Realistic Errir: ',rmse)

result = float(input("Enter the study hrs"))

new_predict = model.predict([[result]])
print(f'Predict score for {result} hours;',new_predict)


```
