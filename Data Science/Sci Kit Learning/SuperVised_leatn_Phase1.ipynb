```python
# linear regression
import pandas as np 
from sklearn.linear_model import LinearRegression


x = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]
model = LinearRegression()
model.fit(x,y)

# how much hrs study
hrs = float(input("enter how many hrs ypou stuided :- "))

predict_hr = model.predict([[hrs]])
print(f'Based on your hours {hrs} you may scored {predict_hr} ')
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

x = [[1],[2],[3],[4],[5]]  # hr studied input
y=[0,0,1,1,1]
model = LogisticRegression()
model.fit(x,y)
hr = float(input("enter hrs many hrs U studied"))
result = model.predict([[hr]])[0]

if result == 1:
    print(f'Based on yur hr {hr} u r likely to pass ')
else:
     print(f'Based on yur hr {hr} u r likely to fail ')
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

x=[
    [100,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]
y=[0,0,0,1,1,1]  # apple = 0 & orange = 1
model = KNeighborsClassifier()
model.fit(x,y)

weight = float(input("Enter witght in grams"))
size = float(input("Enter size in cm"))

result = model.predict([[weight,size]])[0]

if result == 0:
    print(f'this is apple ')
else:
     print(f'this is orange')
```

```python
#decision tree 
from sklearn.tree import DecisionTreeClassifier

x= [
    [7,2],
    [8,3],
    [9,8],
    [10,9],
]
y=[0,0,1,1]  # 0-apple, 1-orange

model = DecisionTreeClassifier()
model.fit(x,y)

size=float(input("Enter the size"))
shade=float(input("Enter the color shade(1-10"))

result = model.predict([[size,shade]])[0]

if result == 0:
    print("apple") 
else:
    print("Orange")


```
