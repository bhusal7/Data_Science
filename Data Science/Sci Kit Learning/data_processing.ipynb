```python
import pandas as pd 

# Data Processing

df = pd.DataFrame({
    'Name' : ['Ram','Shyam','Om','Sita','Gita'],
    'Age' : [23,None,44,23,None],
    'Salary' : [50000,60000,70000,None,None]
})
df

print(df.isnull().mean() * 100)
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df
```
